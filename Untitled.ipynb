encoding

```python
numerical=[]
for col in data.columns:
    if data[col].dtype=="float64":
        numerical.append(col)
print(numerical)
```

```python
categorical=[]
for col in data.columns:
    if data[col].dtype=="object":
        categorical.append(col)
print(categorical)
```

```python
for col in data.columns:
    if col in numerical:
        data[col].fillna(data[col].median(), inplace=True)
    else:
        data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
data2 = data.copy(deep=True)
data3 = data.copy(deep=True)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
```

```python
for col in categorical:
    data[col]=encoder.fit_transform(data[col])
```

```python
data2=pd.get_dummies(data2)
```

```python
data2.head()
```

```python
data3=pd.get_dummies(data3, drop_first=True) 
```

```python
data3.columns
```

```python
mydata = data3.copy(deep=True)
del data3
```

```python
print("Mutual Information between 'Packed Cell Volume' and other continuous features:\n")
from sklearn.feature_selection import mutual_info_regression
mi = mutual_info_regression(data[numerical], data["Year"])

midf = pd.DataFrame(mi, index=numerical, columns=["Year"])

midf
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score


from sklearn.feature_selection import mutual_info_classif
```

```python
print("Normalized Mutual Information between 'Genre' and other categorical features: \n\n")
for col in categorical:
    mi = normalized_mutual_info_score(data[col], data["Genre"])
    print(f"Genre X {col}: {mi}\n")
```
